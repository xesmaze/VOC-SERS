```python
# Simulated Raman spectral data for major compounds in each essential oil
# These are synthetic illustrative data, not actual experimental spectra

raman_shift = np.linspace(600, 1800, 600)  # Raman shift range in cm⁻¹

# Simulated Gaussian peaks centered around key known Raman bands

def gaussian(x, center, width=15, height=1):
    return height * np.exp(-0.5 * ((x - center) / width) ** 2)

# Limonene (Mandarin, Sweet Orange, Spearmint)
limonene = (
    gaussian(raman_shift, 645, height=0.6) +
    gaussian(raman_shift, 990, height=1.0) +
    gaussian(raman_shift, 1435, height=0.8) +
    gaussian(raman_shift, 1670, height=0.7)
)

# Menthol (Mint)
menthol = (
    gaussian(raman_shift, 875, height=0.7) +
    gaussian(raman_shift, 1050, height=0.6) +
    gaussian(raman_shift, 1450, height=0.9)
)

# 1,8-Cineole (Eucalyptus)
cineole = (
    gaussian(raman_shift, 1045, height=0.7) +
    gaussian(raman_shift, 1440, height=0.6)
)

# Linalool (Bergamot)
linalool = (
    gaussian(raman_shift, 1125, height=1.0) +
    gaussian(raman_shift, 1375, height=0.7) +
    gaussian(raman_shift, 1640, height=0.6)
)

# Carvone (Spearmint)
carvone = (
    gaussian(raman_shift, 870, height=0.7) +
    gaussian(raman_shift, 1680, height=1.0)
)

# Methyl salicylate (Wintergreen)
methyl_salicylate = (
    gaussian(raman_shift, 785, height=0.6) +
    gaussian(raman_shift, 1030, height=0.7) +
    gaussian(raman_shift, 1600, height=0.9) +
    gaussian(raman_shift, 1715, height=0.8)
)

```
